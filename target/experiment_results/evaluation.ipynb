```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set()
```

```python
# Define csv header
names = ["alg_name",
         "ntd_name",
         "width",
         "v_t",
         "e_tau",
         "v_tau",
         "graph_name",
         "v_g",
         "e_g",
         "measurement_1",
         "measurement_2",
         "measurement_3",
         "measurement_4",
         "measurement_5",
         "measurement_avg"]
```

```python
# Load Experiment results
# adjust file names after changing the output format in experiment.rs

# Running time experiments with fixed nts classes
brute_force_growth_with_e_tau = pd.read_csv("brute_force_brute_force_growth_with_e_tau_results.csv", names=names)
brute_force_growth_with_graph = pd.read_csv("brute_force_brute_force_growth_with_graph_results.csv", names=names)
diaz_serna_thilikos_growth_with_e_tau = pd.read_csv("diaz_serna_thilikos_diaz_serna_thilikos_growth_with_e_tau_results.csv", names=names)
diaz_serna_thilikos_growth_with_graph = pd.read_csv("diaz_serna_thilikos_diaz_serna_thilikos_growth_with_graph_results.csv", names=names)
modified_dp_growth_with_e_tau = pd.read_csv("modified_dp_modified_dp_growth_with_e_tau_results.csv", names=names)
modified_dp_growth_with_graph = pd.read_csv("modified_dp_modified_dp_growth_with_graph_results.csv", names=names)
diaz_serna_thilikos_mixed_instances = pd.read_csv("diaz_serna_thilikos_mixed_combinations_results.csv", names=names)
modified_dp_mixed_instances = pd.read_csv("modified_dp_mixed_combinations_results.csv", names=names)
# Running time experiments with mixed instances

# Merge all measurements for comparison
all_growth_with_e_tau = pd.concat([brute_force_growth_with_e_tau, 
                                   diaz_serna_thilikos_growth_with_e_tau, 
                                   modified_dp_growth_with_e_tau], 
                                  keys=names, ignore_index=True, sort=False)

all_growth_with_graph = pd.concat([brute_force_growth_with_graph, 
                                   diaz_serna_thilikos_growth_with_graph, 
                                   modified_dp_growth_with_graph], 
                                  keys=names, ignore_index=True, sort=False)

all_results = pd.concat([all_growth_with_e_tau,
                                                all_growth_with_graph,
                                                diaz_serna_thilikos_mixed_instances,
                                                modified_dp_mixed_instances], 
                                  keys=names, ignore_index=True, sort=False)
```
